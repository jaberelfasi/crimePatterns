```python
%matplotlib inline
import pysal as ps
import pandas as pd
import numpy as np

from shapely.geometry import Polygon, point
from geopandas import GeoSeries, GeoDataFrame, read_file
from pandas import DataFrame
import matplotlib.pyplot as plt
import geopandas as gpd
```

```python
#NNHAntisocial117 = read_file('CrimeStatOutput/NNH1_2017Antisocia1CrimeCountNNH.shp')
#NNHViolenceAndSex117 = read_file('CrimeStatOutput/NNH1_2017ViolenceAndSex1CrimeCountNNH.shp')
NNHAntisocial117 = read_file('CrimeStatOutput/NNH1_2017Antisocia1NNH.shp')
NNHViolenceAndSex117 = read_file('CrimeStatOutput/NNH1_2017ViolenceAndSex1NNH.shp')

areas = read_file('data/Clipped.shp')
```

```python
def findNNHInArea(nnh, area):
    value_geo = []
    for i, a in area.iterrows():
        for j, n in nnh.iterrows():
            if a['geometry'].intersects(n['geometry']):
                value_geo.append(a['geometry'])
                nnh = nnh.drop([j])
                
    nnh_lsoa = GeoDataFrame()
    nnh_lsoa['geometry'] = GeoSeries(value_geo)
    
    return nnh_lsoa
```

```python
nnhASlsoas=findNNHInArea(NNHAntisocial117, areas)
```

```python
nnhSVlsoas=findNNHInArea(NNHViolenceAndSex117, areas)
```

```python
disolved = read_file('data/disolved.shp')
fig, ax = plt.subplots(figsize=(20,20))
ax = nnhASlsoas.plot(ax=disolved.plot(ax=ax, color="white"))
plt.title('NNHC Antisocial crimes', size = 30)
```
